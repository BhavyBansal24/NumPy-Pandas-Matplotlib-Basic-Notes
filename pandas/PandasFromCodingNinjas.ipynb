## Handling String based Data

```python
# lets add a male/female column for such string case

df["Gender"] = "Female"
df.iloc[0:135:2,6] = "Male"
df.head(15)
```

```python
# the easy way to handle data is such case is maybe to number them like male with 0 and female with 1

# lets make a function for such and see

def f(s):
    if s == "Male":
        return 0
    else:
        return 1

df["sex"] = df.Gender.apply(f)
del df["Gender"]
df.head()
```
