```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./terrorismData.csv")
#finding frequency for red corridor
df
```

```python
df1 = df[df.State.isin(["Jharkhand","Odisha","Andhra Pradesh","Chhattisgarh"])]
df1.reset_index(inplace=True,drop=True)
df1.dropna()
df1
```

```python
total_killed = df1.Killed.sum()
total_Wounded = df1.Wounded.sum()
total_Casuality = total_killed + total_Wounded
total_year = df1.Year.max() - df1.Year.min()
frequency_RedCorridor = total_Casuality/total_year
print(frequency_RedCorridor) ## hopefully this is correct answer
```
