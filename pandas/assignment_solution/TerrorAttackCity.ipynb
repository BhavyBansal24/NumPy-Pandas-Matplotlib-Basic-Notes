```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./terrorismData.csv")
df.head()
```

```python
df1 = df[df.State == "Jammu and Kashmir"]
df1.reset_index(inplace=True, drop=True)
df1
```

```python
df1.City.mode()
```

```python
df2 = df1[df1.City == "Srinagar"]
df2.reset_index(inplace=True, drop = True)
df2
```

```python
df2.index[df2.Group == "Unknown"]
```

```python
df3 = df2.drop(df2.index[df2.Group == "Unknown"])
df3.reset_index(inplace=True, drop=True)
df3.tail()
```

```python
# now we know srinagar is most Dangerous city & 351 is number of attacks in srinagar
# lets find the most active group 

df3.Group.mode()
```

```python
# hence group with most attacks in srinagar is Muslim Separatists
# lets try to find out the no of attacks by this group in srinagar

df4 = df3[df3.Group == "Muslim Separatists"]
df4.reset_index(inplace=True, drop=True)
df4.tail()
```

```python
# i.e, no of attacks by Muslim Separatists in srinagar is 55
```
