```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./amazon_jobs_dataset.csv")
df.columns =['Index', 'Title', 'location', 'Posting_date', 'DESCRIPTION','BASIC QUALIFICATIONS', 'PREFERRED QUALIFICATIONS']
df.head()
df.drop('Index', axis = 1, inplace=True)
df = df.iloc[df.index[::-1]]
df.reset_index(inplace=True,drop = True)
df.head()
```

```python
new = df["Posting_date"].str.split(",", n = 1, expand = True)
new.head()
```

```python
df["Posting_year"] = new[1]
df.drop("Posting_date", axis = 1, inplace= True)
df = df[['Posting_year']]
df
```
